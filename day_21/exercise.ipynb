```python
class Statistics:
  def __init__(self, sample):
    self.sample = sample
  
  def count(self):
    return len(self.sample)
  def sum(self):
    total = 0
    for item in self.sample:
      total += item
    return total
  def mean(self):
    return self.sum() / self.count()
  def median(self):
    n = self.count()
    sorted_sample = sorted(self.sample)
    if n % 2 == 1:
      return sorted_sample[n // 2]
    i = n // 2
    return (sorted_sample[i - 1] + sorted_sample[i]) / 2
  def mode(self):
    counts = {}
    for item in self.sample:
      counts.setdefault(item, 0)
      counts[item] += 1
    max_count = max(counts.values())
    return [k for k, v in counts.items() if v == max_count]
  def var(self):
    total = sum((item - self.mean()) ** 2 for item in self.sample)
    return total / (self.count() - 1)
  def std(self):
    return self.var() ** 0.5
  def freq_dist(self):
    counts = {}
    for item in self.sample:
      counts.setdefault(item, 0)
      counts[item] += 1
    return sorted(counts.items(), key=lambda x: x[1], reverse=True)
  def min(self):
    return min(self.sample)
  def max(self):
    return max(self.sample)
  def range(self):
    return self.max() - self.min()
  def describe(self):
    print(f'Count: {self.count()}')
    print(f'Sum: {self.sum()}')
    print(f'Min: {self.min()}')
    print(f'Max: {self.max()}')
    print(f'Range: {self.range()}')
    print(f'Mean: {self.mean()}')
    print(f'Median: {self.median()}')
    print(f'Mode: {self.mode()}')
    print(f'Standard Deviation: {self.std():.2f}')
    print(f'Variance: {self.var():.2f}')
    print(f'Frequency Distribution: {self.freq_dist()}')
  
  
ages = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27, 27, 24, 32, 33, 27, 25, 26, 38, 37, 31, 34, 24, 33, 29, 26]


data = Statistics(ages)
print(data.describe())
```

```python
class PersonAccount:
    def __init__(self, first_name, last_name, income, expenses):
        self.first_name = first_name
        self.last_name = last_name
        self.income = income
        self.expenses = expenses

    def total_income(self):
        return sum(self.income)

    def total_expenses(self):
        return sum(self.expenses)

    def account_info(self):
        print(f"Name: {self.first_name} {self.last_name}")
        print(f"Income: {self.total_income()}")
        print(f"Expenses: {self.total_expenses()}")
        print(f"Account Balance: {self.total_income() - self.total_expenses()}")

    def add_income(self, amount):
        self.income.append(amount)

    def add_expense(self, amount):
        self.expenses.append(amount)

    def account_statement(self):
        print(f"Name: {self.first_name} {self.last_name}")
        print(f"Income: {self.total_income()}")
        print(f"Expenses: {self.total_expenses()}")
        print(f"Account Balance: {self.total_income() - self.total_expenses()}")
        print(f"Income: {self.income}")
        print(f"Expenses: {self.expenses}")

    def describe(self):
        print(f"Total Income: {self.total_income()}")
        print(f"Total Expenses: {self.total_expenses()}")
        print(f"Account Balance: {self.total_income() - self.total_expenses()}")
        print(f"Income: {self.income}")
        print(f"Expenses: {self.expenses}")
        
        

income = [1000, 2000, 3000]
expenses = [500, 1000]
person = PersonAccount("John", "Doe", income, expenses)
print(person.describe())
```
