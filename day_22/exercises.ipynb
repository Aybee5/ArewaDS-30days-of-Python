```python
import requests
from bs4 import BeautifulSoup

url = "http://www.bu.edu/president/boston-university-facts-stats/"

r = requests.get(url)
soup = BeautifulSoup(r.text, "html.parser")

boston_data = soup.find_all("ul")

# make the data serializable 
boston_data = [str(item) for item in boston_data]
print(boston_data)
import json
with open("boston-data.json", "w") as f:
    json.dump(boston_data, f)
```

```python
url = "https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States"

r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

presidents_table = soup.select_one(".wikitable").text

# get the data from the table and store in json

with open("presidents-data.json", "w") as d:
    json.dump(presidents_table, d)
print((presidents_table))
```
